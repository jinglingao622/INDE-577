# K-Means Clustering Algorithm

K-Means is an unsupervised machine learning algorithm used for clustering data into a set number of distinct non-overlapping subgroups.

## Characteristics

- **Centroids-Based:** Clusters data around 'k' centroids.
- **Iterative Approach:** Alternates between assigning data to clusters and updating centroids.
- **Euclidean Distance:** Uses distance measures to assign data to the nearest cluster.

## Working Principle

1. **Initialization:** Define 'k' clusters and randomly initialize their centroids.
2. **Assignment:** Allocate each data point to the nearest centroid.
3. **Update:** Recalculate the centroids as the mean of assigned data points.
4. **Convergence:** Repeat until centroids stabilize.

## Applications

- Ideal for applications like market segmentation, document clustering, and image segmentation.

## Limitations

- Must specify the number of clusters beforehand.
- Sensitive to the initial placement of centroids and the presence of outliers.


```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.cluster import KMeans
# Set theme 
sns.set_theme()

from sklearn.datasets import make_blobs
df = pd.read_csv('~/Downloads/Mall_Customers.csv')
X = df.loc[:, ["Annual Income (k$)", "Spending Score (1-100)"]]
X.head()
```

```python
kmeans = KMeans(n_clusters=4)
X["Cluster"] = kmeans.fit_predict(X)
X["Cluster"] = X["Cluster"].astype("category")

X.head()
```

```python
sns.relplot(
    x="Annual Income (k$)", y="Spending Score (1-100)", hue="Cluster", data=X, height=6,
);
```
